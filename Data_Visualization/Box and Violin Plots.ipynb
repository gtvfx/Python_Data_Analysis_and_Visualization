```python
import numpy as np
from numpy.random import randn
import pandas as pd

# stats library
from scipy import stats

# plotting modules
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# iPython command to make the plots appear in the notebook
%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot([data1,data2])
```

```python
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.boxplot(data1,vert=False)
```

```python
#Normal dist
data1 = stats.norm(0,5).rvs(100)

# Tow gamma dist. Concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)])

# Box plot both data1 and data2
sns.boxplot([data1, data2],whis=np.inf)
```

```python
sns.violinplot([data1,data2])
```

```python
sns.violinplot(data2,bw=0.01)
```

```python
sns.violinplot(data1,inner='stick')
```
